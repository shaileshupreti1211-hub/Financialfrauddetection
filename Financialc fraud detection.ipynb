# Financial Fraud Prediction

```python
 import numpy as np
 import pandas as pd
 from sklearn.model_selection import train_test_split
 import matplotlib.pyplot as plt
 import seaborn as sns
 from sklearn.ensemble import RandomForestClassifier
 from sklearn.metrics import confusion_matrix, classification_report
```

```python
p1= pd.read_csv('creditcard.csv')
```

```python
p1
```

```python
p1.info()
```

## Checking for null values

```python
p1.isnull().sum()
```

## Total Number of legit and fraud transactions

```python
legit_count = len(p1[p1['Class'] == 0])
fraud_count = len(p1[p1['Class'] == 1])
print (f"Normal transactions: {legit_count}")
print (f"Fraud cases: {fraud_count}")
```

# Creating pie chart for the fradulant and normal transaction

```python
 class_counts = p1['Class'].value_counts()

 # Create pie chart

 plt.figure(figsize=(5, 3))
 plt.pie(class_counts,
 labels=['Legitimate (0)', 'Fraudulent (1)'],
 colors=['yellow', 'red'],
 autopct='%1.1f%%',
 startangle=90,
 explode=(0, 0.1))
 plt.title('Class Distribution (0 vs 1)')
 plt.axis('equal')
 plt.show()
```

# Prediction

```python
 X = p1.drop(['Class', 'Time'], axis=1)
 y = p1['Class']
```

```python
 from sklearn.preprocessing import StandardScaler
 scaler = StandardScaler()
 X_scaled = scaler.fit_transform(X)
```

# Test, train and split

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.1,
 random_state=42, stratify=y)
```

#  Create the Random Forest classification model as a result of high imbalance between fradulant(0.2%) and legitmate (99.8%) transactions

```python
rf_model = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
```

```python
rf_model
```

# Fitting training data into the model

```python
rf_model.fit(X_train, y_train)
```

# Prediction and evaluation

```python
y_pred = rf_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred, digits=4))
```

```python
 # Get and sort feature importances
 importances = rf_model.feature_importances_
 sorted_idx = importances.argsort()
 plt.figure(figsize=(10, 8))
 plt.barh(X.columns[sorted_idx], importances[sorted_idx],
 color='#1f77b4', edgecolor='none', height=0.7)
 plt.title('Top Fraud-Detecting Factors', pad=20, fontsize=14, fontweight='bold')
 plt.xlabel('Importance Score', fontsize=12)
 plt.gca().spines['top'].set_visible(False)
 plt.gca().spines['right'].set_visible(False)
# Add value labels
 for i, v in enumerate(importances[sorted_idx]):
     plt.text (v + 0.001, i, f"{v:.3f}", color='black', ha='left', va='center', fontsize=10)
 plt.tight_layout()
 plt.show()
```
